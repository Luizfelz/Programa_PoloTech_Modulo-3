1) Utilizando o DataFrame importado (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)

```python
import pandas as pd
# OBS: ajustar o 'path' de acordo com o local onde está salvo o arquivo alunos3.csv
df = pd.read_csv('/content/alunos3.csv', sep=';', decimal=',') # lê o arquivo.
df.sort_values(by=['Nome'],inplace=True) # ordena o dataframe em ordem alfabética
df.reset_index(drop=True,inplace=True) # reseta o index, apagando o anterior
colunas = list(df.columns) # cria uma lista com os nomes das colunas
new_df = pd.melt(df, id_vars=colunas[0:3],value_vars=colunas[3:],var_name='Prova',value_name='Nota') # dá 'melt' no dataframe, unpivotando somente as colunas com as notas das provas
new_df.sort_values(by=['Nome','Prova'],inplace=True) # ordena o novo dataframe em ordem alfabética e pela ordem das provas
new_df.reset_index(drop=True,inplace=True) # reseta o index, apagando o anterior
Media_provas = [] # lista vazia onde ficarão armazenados os valores das médias ddas notas das provas
k = 0 # variável temporária
for i in range(df.shape[0]):
  soma = 0
  nome_temp = df.iat[i,1] # atribui o primeiro nome do primeiro aluno (ordem alfabetica) à variável temporária. O índice 1 representa a coluna de 'Nome'
  while(nome_temp == new_df.iat[k,1]):
    soma += new_df.iat[k,-1]
    k += 1
    if k == new_df.shape[0]: # finaliza o loop após percorrer todos os nomes
     break
  Media_provas.append(soma/4)
df['Media_provas'] = Media_provas # acrescenta a coluna Media_provas ao dataframe original
df.head(10)
```

2) Quem foram os alunos que obtiveram a maior e a menor média

```python
maior_media = max(Media_provas)
index_maior_media = Media_provas.index(maior_media)
menor_media = min(Media_provas)
index_menor_media = Media_provas.index(menor_media)

print(f'O aluno(a) com a maior média é: {df.iat[index_maior_media,1]}, cuja média = {maior_media}.')
print(f'O aluno(a) com a menor média é: {df.iat[index_menor_media,1]}, cuja média = {menor_media}.')
```

3) Agora una este dataframe com o `cadastro_alunos.xlsx`

```python
# OBS: ajustar o 'path' de acordo com o local onde está salvo o arquivo cadastro_alunos.xlsx
df2 = pd.read_excel('/content/cadastro_alunos.xlsx')
df.sort_values(by=['RA'],inplace=True)
merged_df = df.merge(df2,how='left',on=['RA']) # junta os dataframes, trazendo as colunas do df2, igualando-os com base na coluna de nome 'RA'. Ambos devem ter uma coluna com mesmo nome para que funcione o .merge()
merged_df.head(15)
```

4) Qual a média entre as Media_provas dentro do público feminino? e masculino?

```python
df_feminino = merged_df[merged_df['Sexo'] == 'Feminino'] # cria um novo dataframe filtrando o público feminino
df_masculino = merged_df[merged_df['Sexo'] == 'Masculino'] # cria um novo dataframe filtrando o público masculino
media_masculino = df_masculino['Media_provas'].mean()
media_feminino = df_feminino['Media_provas'].mean()

print(f'A média das notas para o público feminino é = {media_feminino}.')
print(f'A média das notas para o público masculino é = {media_masculino}.') # machista!!!  ksksksksksksk   Zueira  :P
```

5) Qual a média de idade das pessoas que obtiveram Media_provas maior ou igual a 7?

```python
df_media_maior_7 = merged_df[merged_df['Media_provas'] >= 7]
media_idade = df_media_maior_7['Idade'].mean()

print(f'A média das idades dos alunos que tiveram nota >= 7 é = {media_idade:.2f} anos.')
```

6) Qual das cidades possui o maior média de Media_provas? E qual é este valor?

```python
df_cidades = merged_df.pivot_table(columns='Cidade',values='Media_provas', aggfunc='mean') # cria um dataframe tipo pivot_table com as cidades e as médias, por cidade, mas Media_provas
medias_por_cidade = list(df_cidades.loc['Media_provas']) # cria uma lista com as médias por cidade
maior_media_por_cidade = max(medias_por_cidade) # pega o maior valor da média por cidade
index_maior_media_por_cidade = medias_por_cidade.index(maior_media_por_cidade) # retorna o índice da maior média por cidade
cidade_maior_media = df_cidades.columns[index_maior_media_por_cidade] # retorna o nome da cidade com maior média, de acordo com o índice da média em si

print(f'A cidade com a maior média é {cidade_maior_media} e a média é = {maior_media_por_cidade}.')
```
